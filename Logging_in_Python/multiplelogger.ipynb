```python
import logging
```

```python
# creating a logger for module 1

logger1=logging.getLogger("module1")
logger1.setLevel(logging.DEBUG)

# logger for module 2

logger2=logging.getLogger("module2")
logger2.setLevel(logging.WARNING)

# configuring logging settings

logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
)
```

```python
#  log message with different loggers

logger1.debug("this is a debug message for module 1")
logger2.warning('this is a warning message for mo')
```

```python
logger1.error("this is a error message for module 1")
logger2.critical('this is a critical message for mo')
```
