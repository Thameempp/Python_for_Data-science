```python
import threading
import time
```

```python
def print_numbers():
    for i in range(5):
        time.sleep(2)
        print(f"number:{i}")

def print_letter():
    for letter in 'abcde':
        time.sleep(1.5)
        print(f"Letter:{letter}")

th1=threading.Thread(target=print_numbers)
th2=threading.Thread(target=print_letter)

t=time.time()

th1.start()
th2.start()


th1.join()
th2.join()

finished_time = time.time() - t
print(finished_time)
```
